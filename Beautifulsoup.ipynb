# BeautifulSoup

```python
##Import the required Libraries.
import bs4 as bs
import urllib.request
import pandas as pd
```

```python
#Open the URL and storing in a variable called sauce.
sauce=urllib.request.urlopen("http://www.srikanthtechnologies.com/rss.xml").read()
```

```python
#Parsing the url information and using BeautifulSoup library and storing in variable soup
soup=bs.BeautifulSoup(sauce,'xml')
```

```python
#Getting info from title attribute and store them in titles which is a list variable.
titles=[]
for item in soup.findAll('item'):
    for title in item.findAll('title'):
        text_obj=title.getText()
        titles.append(text_obj)
```

```python
#Getting info from link attribute and store them in linkss which is a list variable.
linkss=[]
for item in soup.findAll('item'):
    for link in item.findAll('link'):
        text_link=link.getText()
        linkss.append(text_link)
        
```

```python
#Getting info from description attribute and store them in descriptionss which is a list variable.
descriptionss=[]
for item in soup.findAll('item'):
    for desc in item.findAll('description'):
        text_desc=desc.getText()
        descriptionss.append(text_desc)
```

```python
#Getting info from Date attribute and store them in Datess which is a list variable.
Datess=[]
for item in soup.findAll('item'):
    for date in item.findAll('pubDate'):
        text_date=date.getText()
        Datess.append(text_date)
        
```

```python
#keeping everything into individual DataFrames as follows.
Linksdf=pd.DataFrame(linkss,columns=['Links'])

Titlesdf=pd.DataFrame(titles,columns=['Titles'])

Descdf=pd.DataFrame(descriptionss,columns=['Description'])

Datesdf=pd.DataFrame(Datess,columns=['Dates'])
```

```python
#Combining all the individual DataFrames into a single one.
MainFrame=pd.concat([Titlesdf,Descdf,Linksdf,Datesdf],axis=1)
```
